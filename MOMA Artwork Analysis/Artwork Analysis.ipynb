Create function that cleans a date input

```python
def clean_and_convert(date):
    #date = date.replace(np.nan, 0)
    #date = str(date)
    
        date = str(date)
        
    
    # check if date includes a preceding hyphen
    if date[0] == '-':
        date = date[1:]
        
    # check if date represents a range of years
    if '-' in date:
        # if the value is a range of years, split the values by the hyphen
        mult_years = date.split('-')
        year_sum = 0
        # loop through multiple years
        for year in mult_years:
            year = int(year)
            year_sum += year
        # calculate average between beginning and end year of a range    
        date = round(year_sum / 2)
        
    # check if date has parentheses
    if date != "":
        # remove parentheses
        date = date.replace("(", "")
        date = date.replace(")", "")
        
    # check if date includes a space
    if ' ' in date:
        # if the value is a range of years, split the values by the hyphen
        mult_years = date.split(' ')
        if mult_years[0] == '0':
            date = mult_years[1]
        else:
            date = mult_years[0]
    
    if len(date) > 4:
        date = date[:4]
        
    # convert date to int 
    if date == '':
        date = 0
    else:
        date = int(date)
        
    # return clean date
    return date
```
