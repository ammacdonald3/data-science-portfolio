#### Import Libraries

```python
import pandas as pd
```

```python
crime_df = pd.read_csv('crime.csv', encoding = "ISO-8859-1")
```

```python
print(crime_df.head())
```

```python
pd.DataFrame(crime_df.OFFENSE_CODE_GROUP.unique())
```

```python
crime_count_df = crime_df.groupby(['YEAR', 'OFFENSE_CODE_GROUP'])['INCIDENT_NUMBER'].count().reset_index().sort_values(by=['YEAR', 'INCIDENT_NUMBER'], ascending=False)
```
