### Pandas Loops

In this exercise, we're going to produce a dictionary of the top employer in each country

```python
#pd.set_option("display.max_rows", None, "display.max_columns", None)
top_employer_by_country = {}

countries = f500['country'].unique()

for country in countries:
  selected_countries = f500[f500['country'] == country]
  sorted_countries = selected_countries.sort_values('employees', ascending=False)
  first_company_row = sorted_countries.iloc[0]
  first_company_name = first_company_row['company']
  top_employer_by_country[country] = first_company_name

top_employer_by_country
```

### Return on Assets

Create a new column roa in the f500 dataframe, containing the return on assets metric for each company.

```python
roa = f500['profits'] / f500['assets']

f500['roa'] = roa

top_roa_by_sector = {}

sectors = f500['sector'].unique()

for sector in sectors:
  selected_companies = f500[f500['sector'] == sector]
  sorted_companies = selected_companies.sort_values('roa', ascending=False)
  top_company = sorted_companies.iloc[0]
  top_company_name = top_company['company']
  top_roa_by_sector[sector] = top_company_name

top_roa_by_sector
```
